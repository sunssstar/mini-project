```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('교통_2019_연령24세이하_남성.csv')
```

```python
df1.head()
```

```python
df1.sort_values(by='가계지출_소비지출_교통_육상운송금액',ascending=False)
```

```python
열별_평균 = df1.mean()
print(열별_평균)
```

```python
a = '가계지출_소비지출_교통비'
평균 = df1[a].mean()
중앙값 = df1[a].median()
최고값 = df1[a].max()
최소값 = df1[a].min()
x = round(평균,-2)

y = round(중앙값,-2)

z = round(최고값,-2)

r = round(최소값,-2)
print(a)
print(f"최고값: {z}원")
print(f"최소값: {r}원")
print(f"평균: {x}원")
print(f"중앙값: {y}원")
```
